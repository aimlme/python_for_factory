# Matplotlib

การใช้งานพื้นฐานของ Matplotlib สำหรับการสร้างกราฟประเภทต่างๆ

https://matplotlib.org


```python

import numpy as np
import matplotlib.pyplot as plt
```

## 1. กราฟเส้น (Basic Line Plot)
ตัวอย่างนี้แสดงวิธีสร้างกราฟเส้นง่ายๆ ด้วย `matplotlib.pyplot`

```python

X = np.linspace(0, 2*np.pi, 100)
Y = np.cos(X)

fig, ax = plt.subplots()
ax.plot(X, Y, color='green')

plt.show()
```

## 2. องค์ประกอบของกราฟ (Figure Anatomy)
ตัวอย่างนี้แสดงการเพิ่มป้ายชื่อแกน, ชื่อกราฟ, grid และคำอธิบายเส้นกราฟ

```python

fig, ax = plt.subplots()
ax.plot(X, Y, label='cos(x)', color='blue', linestyle='--', marker='o')

ax.set_title("cosine graph")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()
ax.grid(True)

plt.show()
```

## 3. การบันทึกกราฟเป็นไฟล์ (Saving Figures)
ใช้ `.savefig()` เพื่อบันทึกกราฟเป็นไฟล์

```python

fig, ax = plt.subplots()
ax.plot(X, Y, color='red')

fig.savefig("figure.pdf")  # บันทึกกราฟเป็นไฟล์ PDF
plt.show()
```

## 4. การสร้างหลายกราฟในรูปเดียว (Multiple Subplots)
ใช้ `plt.subplots()` เพื่อจัดเรียงกราฟหลายๆ กราฟไว้ในรูปเดียว

```python

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(X, np.sin(X), color='blue')
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(X, np.cos(X), color='green')
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(X, np.tan(X), color='red')
axs[1, 0].set_title("Tangent Wave")

axs[1, 1].plot(X, -np.sin(X), color='purple')
axs[1, 1].set_title("Negative Sine Wave")

plt.tight_layout()
plt.show()
```

## 5. ประเภทของกราฟพื้นฐาน (Basic Plot Types)
Matplotlib รองรับการสร้างกราฟหลายประเภท เช่น กราฟเส้น, กราฟกระจาย, กราฟแท่ง, ฮิสโตแกรม, และพายชาร์ต

```python

# กราฟเส้น (Line Plot)
plt.figure(figsize=(6,4))
plt.plot(X, np.sin(X), label="Sine", color='blue')
plt.plot(X, np.cos(X), label="Cosine", color='red')
plt.legend()
plt.title("example line plot")
plt.show()

# กราฟกระจาย (Scatter Plot)
plt.figure(figsize=(6,4))
plt.scatter(X, np.sin(X), color='blue', label="Sine")
plt.scatter(X, np.cos(X), color='red', label="Cosine")
plt.legend()
plt.title("example scatter plot")
plt.show()

# กราฟแท่ง (Bar Chart)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.figure(figsize=(6,4))
plt.bar(categories, values, color=['blue', 'red', 'green', 'purple'])
plt.title("example bar chart")
plt.show()

# ฮิสโตแกรม (Histogram)
data = np.random.randn(1000)
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, color='gray', edgecolor='black')
plt.title("example histogram")
plt.show()
```
