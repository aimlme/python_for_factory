```python
import requests
import cv2
import numpy as np
from PIL import Image
from io import BytesIO
from google.colab.patches import cv2_imshow

# Use a reliable image URL
image_url = "https://images.dmc.tv/wallpaper/raw/4319.jpg"

try:
    # Fetch the image
    response = requests.get(image_url)
    response.raise_for_status()  # Raise error for bad responses

    # Convert image content to numpy array
    image = np.array(Image.open(BytesIO(response.content)))

    # Convert RGB to BGR for OpenCV
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Print shape
    print("Image shape:", image.shape)

    # Display image in Google Colab
    cv2_imshow(image)

except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
except Exception as e:
    print(f"Error processing image: {e}")
```

```python

from matplotlib import pyplot as plt

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Plot downloaded image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Downloaded Picture')

# Plot gray image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB))
plt.axis('off')
plt.title('Grays Picture')

plt.show()
```
